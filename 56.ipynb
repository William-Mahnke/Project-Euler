## Powerful Digit Sum

A googol ($10^100$) is a massive number: one followed by one-hundred zeros; 100^100 is almost unimaginably large: one followed by two-hundred zeros. Despite their size, the sum of the difits in each number is only 1.

Considering natural numbers of the form, $a^b$, where $a, b < 100$, what is the maximum digit sum?

### Methodology 

Using a nested for loop (which does have $O(n^2)$ complexity), we can iterate through the combinations. For each iteration we will have to convert the integer into a string then into a list of charaters and then calculate the digit sum.

```python
# nested loop
for a in range(1, 101):
    for b in range(1, 101):
        val = a ** b
        val_str = str(val)
```
