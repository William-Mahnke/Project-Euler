## Largest Product in Grid

What is the greatest product of four adjacent numbers in the same direction (up, down, left, right, or diagonally) in the following 20x20 grid?

NOTE: didn't display the grid here because of difficulty in formatting the grid

### Methodology

Solving this problem can be done iterating through the entries in the grid. Since going up and down or left and right can produce the same result (e.g. doing the product of four numbers in the downward direction would be the same result as the product of the four numbers in the up direction starting at the lower point), we can save time by starting at the top and working our way down (going right, down, diagonally left, and diagonally right). The only other thing to be wary of is the indexing, making sure to not go outside the index of the matrix.

To ease the transformation of the grid of numbers into a matrix I can use, I'll create a numpy array. It's also perfectly viable to use nested lists with the same approach, I just used numpy to make indexing easier.

```python
# iterate through rows and columns keeping track of product
# variable for largest product
largest = 0

for i in range(20):
    for j in range(20):
        # can speed up the process slightly
        # if the number we're currently at is 0, we can iterate to the next number
        if grid[i, j] == 0:
            continue
        
        # if the first number isn't a zero we can proceed with
        # calculating the products in different directions
        # for each prod we calculate we also have to check if it's 
        # larger than largest (and reassign largest if so)

        # start with to the right
        # check if i is less than 17 (otherwise product would go out of range)
        if j < 17:
            prod = grid[i, j] * grid[i, j+1] * grid[i, j+2] * grid[i, j+3]
            if prod > largest:
                largest = prod
        
        # going down
        # check if j is less than 17 (otherwise product would go out of range)
        if i < 17:
            prod = grid[i, j] * grid[i+1, j] * grid[i+2, j] * grid[i+3, j]
            if prod > largest:
                largest = prod

        # going diagonal down to the left
        # check if i is greater than 2 and j is less than 17
        if i > 2 and j < 17:
            prod = grid[i, j] * grid[i-1, j+1] * grid[i-2, j+2] * grid[i-3, j+3]
            if prod > largest:
                largest = prod
        
        # going diagonal down to the right
        # check if i is less than 17 and j is less than 17
        if i < 17 and j < 17:
            prod = grid[i, j] * grid[i+1, j+1] * grid[i+2, j+2] * grid[i+3, j+3]
            if prod > largest:
                largest = prod

# return largest product
print(f'Largest product from the grid: {largest}')
```
