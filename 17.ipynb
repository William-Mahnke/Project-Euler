## Number Letter Counts

If the numbers 1 to 5 are written out in words: one, two, three, four, five, then there are $3+3+5+4+4=19$ letters used in total.

If all the numbers from 1 to 1000 (one thousand) inclusive were written out in words, how many letters would be used?

NOTE: Don't count spaces or hypens. For example, 342 (three hundred and forty-two) contains 23 letters and 115 (one hundred and fifteen) contains 20 letters. The use of "and" when writing out numbers is in compliance with British usage.

### Methodology

Coding isn't necessary for solving this problem. We can break the numbers into different portions.

Firstly, we can calculate how many letters will be used for the 1-99 part of all numbers (from 1 to 999). For 1-9, the number of letters is 36. For 10-19, the number of letters used is 70. For 20-99 we can add the number of times each tens place (twenty, thirty, etc.) with the letters used for the ones place in each set. Each word in the tens place (twenty, thirty, ...) will be used ten times so we can multiply the sum of the words' digits, 46, by 10 to get 460. Additionally each set of numbers (20-29, 30-39, ...) will use the letters for 1-9. Since there are 8 sets of these, the number of letters used for this will be 36 * 8 = 288. Thus the number of letters 20-99 will be 460 + 288 = 748. Therefore the number of letters used for 1-99 is 36 + 70 + 748 = 854. Since these letters will be used for each hundred, we can multiply 854 by 10 to get 8540 to get the number of letters used for the 1-99 portion of all numbers from 1-1000. 

Second, we need to calculate the letters used for the hundreds place which will also be broken down into different parts. For each multiple of 100 the letters used will be the digits 1-9 plus the word "hundred". Saving the digit part for later we can multiply the letters used for hundred, 7, times 9 to get 63. For the numbers which aren't multiples of 100 we use "hundred and". For each set of hundred numbers (100-199, 200-299, ...) these ten letters will be used 99 times. Since this applies for 9 groups we can multiply 10 by 99 and then by 9 to get 8910. Finally we can look at the letters used for the digits in the hundreds place. Since the letters for the digit will be used 100 times each, we can take the sum of the letters for 1-9, 36, and multiply that 100 to get 3600. 

Adding in the 11 letters for "one thousand", we get the total number of digits used as:

8540+8910+3600+63+11=21124
