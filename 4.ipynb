## Largest Palindrome Product

A palindromic number reads the same both ways. The largest palindrome made from the product of two 2-digit numbers is $9009 = 91 \times 99$. 

Find the largest palindrome made from the product of two 3-digit numbers. 

### Methodology

Using a nested for loop for the two three-digit numbers, we can find each product and then check if it's palindromic. If it is, and it's also larger than the current largest product, we assign the variable a new value and proceed with the next iteration. 

```python
# variable for largest product
largest = 0

# nested loop
for a in range(100, 1000):
    for b in range(100, 1000):
        product = a * b

        # can save time by first checking if product > largest 
        # before checking if it's palindromic
        if product > largest:
            # now check if palindromic
            if str(product) == str(product)[::-1]:
                # number is palindromic and larger than largest
                # thus, reassign largest with product
                largest = product

# print result after loops
print(f'Largest palindromic product: {largest}')
```
