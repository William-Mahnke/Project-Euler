## Largest Product in a Series

Find the thirteen adjacent digits in the 1000-digit number that have the greatest product. What is the value of this product?

### Methodology 

Since the number is long, we can first convert the number to a list of characters (after turning the integer into a string) and convert the values to numbers. Then, we can iterate through the list keeping track of the largest product as we go.

```python
# create list for numbers and iterate through the initial large number
digits = []

for digit in num:
    if digit == "\n":
        continue
    digits.append(int(digit))
```

```python
# iterate through list and keep track of largest product
largest = 0

for i in range(988):
    prod = np.prod([digits[j] for j in range(i, i + 13)])

    if prod > largest: # new largest product
        largest = prod

# print largest product
print(f'Largest Product: {largest}')
```
