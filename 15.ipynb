## Lattice Paths

Starting in the top left corner of a $2 \times 2$ grid, and only being able to move to the right and down, there are exactly 6 routes to the bottom right corner.

How many such routes are there through a $20 \times 20$?

### Methodology

There is no code needed to solve this problem. If we look at the $2 \times 2$ example presented in the problem, we can represent each path as a string of R's (to represent moving right) and D's (to represent moving down). For each path the total number of moves is 4, the sum of the length and width of the grid. Additionally, each path requires moving right exactly two times and moving down exactly two times. Therefore each string for a path will be a combination of 2 R's and 2 D's. If we consider the string with respect to the R's, we can can calculate the number of combinations by considering the number of ways we can place two R's in a string of 4 letters (since the D's must fill the remaining spots). With 4 possible slots for 2 letters the number of possibilities is 4C2 (four choosing two) which equals 6. Applying this to the $20 \times 20$ grid changes the path length to 40 and the number of R's to 20. Thus the total number of possible paths is 40C20 = 137846528820.
