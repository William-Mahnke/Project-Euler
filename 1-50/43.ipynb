## Sub-string Divisibility

The number, 1406357289, is a 0 to 9 pandigital number because it is made up of each of the digits 0 to 9 in some order, but it also has a rather interesting sub-string divisibility property.

Let $d_1$ be the 1st digit, $d_2$ be the 2nd digit, and so on. In this way, we note the following:

* $d_2d_3d_4 = 406$ is divisible by 2
* $d_3d_4d_5 = 063$ is divisible by 3
* $d_4d_5d_6 = 635$ is divisible by 5
* $d_5d_6d_7 = 357$ is divisible by 7
* $d_6d_7d_8 = 572$ is divisible by 11
* $d_7d_8d_9 = 728$ is divisible by 13
* $d_8d_9d_{10} = 289$ is divisible by 17

Find the sum of all 0 to 9 pandigital numbers with this property.

### Methodology

NOTE: for this problem we can assume the first digit won't be 0.

This problem can be done without any code. Looking at the third property, we can see that if $d_4d_5d_6$ must be divisible by 5 then $d_6$ will either be 0 or 5. This means if $d_6d_7d_8$ must be divisible by 11 then those three digits must be a number divisible by 11 which also starts with a 0 or a 5. For the case when the first digit is 0 (i.e. two-digit multiples of 11), all of these multiples repeat digits. Since the number is pandigital, we can eliminate the possibility of $d_6 = 0$ and proceed with the multiples of 11 in the 500s. 

The three-digit multiples of 11 which start with a 5 are as follows: 506, 517, 528, 539, 550, 561, 572, 583, 594. We can first remove 550 because it repeats digits. Then for the remaining possibilities we can fill in digits to see if there are possible pandigital numbers. 

506 $\rightarrow$ _ _ _ _ _ 5 0 6 _ _: If $d_6 = 5, d_7 = 0, d_8 = 0$ then $d_9$ would have to be 5 (065 is the only multiple of 13 which works). Since this digit is already used, we can conclude this choice of digits for spots 6, 7, and 8 won't work. 

517 $\rightarrow$ _ _ _ _ _ 5 1 7 _ _: In this case there is no value for $d_9$ which will satisfy the requirement for $d_7d_8d_9$ (no three-digit multiple of 13 starts with 17). Thus this choice won't work.

528 $\rightarrow$ _ _ _ _ _ 5 2 8 _ _: This choice of digits forces $d_9$ to be 6 and consequently forces $d_{10}$ to be 7. Working backwards and starting with filling $d_5$, the two three-digit multiples of 7 which end with 52 are 252 and 952. Since 252 uses a digit twice, the only possibility is $d_5 = 9$. At this point the number has the form _ _ _ _ 9 5 2 8 6 7. Of the remaining digits (0, 1, 3, and 4), the fact $d_3d_4d_5$ must be divisible by 3 forces $d_3$ and $d_4$ to be 0 and 3. Additionally, since $d_2d_3d_4$ is divisible by 2, we are forced to set $d_3 = 3, d_4 = 0$. With these choices we can interchange 1 and 4 in the first two digits which leads us to two pandigital numbers which satisfies the property: 1430952867, 4130952867.

539 $\rightarrow$ _ _ _ _ _ 5 3 9 _ _: In this case $d_9 = 0$ which means $d_{10} = 1$. Then $d_5 = 5$, but that digit is already being used so this choice won't work.

561 $\rightarrow$ _ _ _ _ _ 5 6 1 _ _: The only digit for $d_9$ which works is 1. Since that digit is already being used, this case won't work.

572 $\rightarrow$ _ _ _ _ _ 5 7 2 _ _: In this case we get $d_9 = 8$ which implies $d_{10} = 9$ which prompts $d_5 = 3$. Thus any candidate number is of the form _ _ _ _ 3 5 7 2 8 9. Of the remaining digits (0, 1, 4, 6) the property thus requires $d_3$ and $d_4$ to be 0 and 6. Since they're both even they can be placed interchangably and the remaining two digits can also be switched. Thus with this case we get four different candidates: 1406357289, 4106257289, 1460357289, and 4160357289.

583 $\rightarrow$ _ _ _ _ _ 5 8 3 _ _: In this case $d_9 = 2$ and $d_{10} = 3$. Since 3 is already being used, this case won't work. 

594 $\rightarrow$ _ _ _ _ _ 5 9 4 _ _: In this case $d_9$ has to be 9 which means this case won't work. 

Therefore, if we add up the six numbers which satisfy the property we get the final solution of 16695334890.
