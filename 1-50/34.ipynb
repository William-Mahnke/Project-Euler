## Digit Factorials 

145 is a curious number, as $1! + 4! + 5! = 1 + 24 + 120 = 145$. 

Find the sum of all numbers which are equal to the sum of the factorial of their digits. Note: As $1! = 1$ and $2! = 2$ are not sums they are not included.

### Methodology

We can brute force this problem by going up to some arbitrary limit, say 1000000, and checking if the criterion is met everytime. Since we're just working with the factorials for single digit numbers, we can save time by saving the values in a dictionary. While it's impossible to verify a good upper limit to check numbers, intuitively it's hard to see the criterion being met for a value after one million.

```python
from math import factorial
```

```python
# dictionary for factorial values
fact = {str(i) : factorial(i) for i in range(0, 10)}
```

```python
# iterate through and calculate total
total = 0
for n in range(10, 1000000):
    if sum(fact[d] for d in str(n)) == n:
        total += n

# print total
print(f'Total for numbers meeting the criterion: {total}')
```
