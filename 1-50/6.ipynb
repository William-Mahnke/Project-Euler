## Sum Square Difference

The sum of the squares of the first ten nautral numbers is: $1^2 + 2^2 + \dots + 10^2 = 385$.

The square of the sum of the first ten natural numbers is: $(1 + 2 + \dots + 10)^2 = 55^2 = 3025$.

Hence the difference of the square of the sum and the sum of the squares is $3025 - 385 = 2640$.

Find the difference of the square of the sum of the first one hundred natural numbers and the sum of the sqaures.

### Methodology 

Calculating the difference requires calculating the two values first. To expedite the process we can use lists and `sum()` to compute the values easily before taking this difference.

```python
# sum of squares
sum_of_sqs = sum([i**2 for i in range(1, 101)])

# square of sum
sq_of_sum = (sum([i for i in range(1, 101)]))**2

# print difference
print(f"Difference between the square of sum and sum of squares: {sq_of_sum - sum_of_sqs}")
```
