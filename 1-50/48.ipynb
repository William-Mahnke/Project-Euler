## Self Powers

The series, $1^1 + 2^2 + 3^3 + \dots + 10^{10} = 10405071317$.

Find the last ten digits of the series, $1^1 + 2^2 + 3^3 + \dots + 1000^{1000}$.

### Methodology

Since all of these numbers can be calculated, we can do a simple loop to add these numbers together. We can then convert the sum to a string to find the last ten digits.

```python
# variable for total
total = 0

# calculate total
for i in range(1, 1001):
    total += (i ** i)

# convert total to string
total_str = str(total)

# print result
print(f'Last ten digits of the series: {total_str[-10:]}')
```
