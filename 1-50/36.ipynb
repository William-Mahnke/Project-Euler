## Double-base Palindromes

The decimal number, $585 = 1001001001_2$ (binary), is palindromic in both bases. Find the sum of all numbers, less than one million, which are palindromic in base 10 and base 2 (no leading zeros).

### Methodology

First we'll go through all numbers less than one million and find the palindromic numbers. Then with that list, we can convert the numbers to base 2 and check if those numbers are palindromes before computing the sum.

```python
# filter for palindromic numbers
palindromes = []

for i in range(1, 1000001):
    str_i = str(i)
    if str_i == str_i[::-1]:
        palindromes.append(i)
```

```python
# check if binary rep of palindromes are palindromes themselves
dbl_palindromes = []

for num in palindromes:
    bin_num = format(num, 'b')
    if bin_num == bin_num[::-1]:
        dbl_palindromes.append(num)

# print sum of dbl_palindromes
print(f"Sum of double-base palindromes: {sum(dbl_palindromes)}")
```
