## Factorial Digit Sum

$n!$ means $n \times n-1 \times \dots \times 3 \times 2 \times 1$. For example, $10! = 10 \times 9 \times \dots \times 3 \times 2  \times 1 = 3628800$, and the sum of the digits in the number is $3 + 6 + 2 + 8 + 8 + 0 + 0 = 27$. Find the sum or the digits in the number $100!$

### Methodology

While we could use a package to calculate the factorial, using an easy function is also easily implementable. From there we can use the same strategy as employed in problem 16, converting the number to a string then to a list and summing the digits. 

```python
# factorial function and calculating 10!
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n-1)

num = factorial(100)
```

```python
# convert to string
num = str(num)

# convert string to list
digits = [int(digit) for digit in num]

# sum digits
print(f'Sum of digits: {sum(digits)}')
```
