It can be seen that the 12th digit of the fractional part is 1.

If $d_n$ represents the nth digit of the fractional part, find the value of the following expression.

$d_1 \times d_{10} \times d_{100} \times d_{1000} \times d_{10000} \times d_{100000} \times d_{1000000}$

### Methodology

Using a while loop, we can add string versions of integers to a string until the length of the string exceeds a million characters. Then we can use indexing to find the product of the appropriate digits. 

```python
dec = ''
i = 1
while len(dec) < 1000000:
    dec += str(i)
    i += 1
```

```python
res = int(dec[0]) * int(dec[9]) * int(dec[99]) * int(dec[999]) * int(dec[9999]) * int(dec[99999]) * int(dec[999999])
print(f'Resulting expression: {res}')
```
