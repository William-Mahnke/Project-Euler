## Pandigital Primes

We shall say that an ð‘›-digit number is pandigital if it makes use of all the digits 1 to ð‘› exactly once. For example, 2143 is a 4-digit pandigital and is also prime.

What is the largest pandigital prime to exist?

```python
# use sympy to find primes faster
import sympy as sp
```

```python
# function to check if pandigital
def pandigital(n:int) -> bool:
    '''
    In put for the function will always be a prime number,
    thus this function only has to check whether the number is
    pandigital.
    '''
    # convert n to string
    n = str(n)
    # pandigital doesn't include using 0, so can filter for that first
    if '0' in n:
        return False
    # dict for storing digit appearances
    counts = {str(i):0 for i in range(1, len(n)+1)}
    for d in n:
        # if a digit greater than len(n) is in the string, n can't be 
        # pandigital
        if d not in counts.keys():
            return False
        counts[d] += 1
    # create set of values from counts
    # if a number is pandigital (only uses each digit exactly once), then
    # set of values from counts should just be {1}
    if set(counts.values()) == {1}:
        return True
    else:
        return False
```

```python
# start from largest prime less than upper limit for pandigital numbers 
# 1000000000 
n = sp.prevprime(1000000000)

# loop to iterate through primes (going backwards)
while True:
    if pandigital(n):
        print(f'Largest n-digit pandigital prime: {n}')
        break
    else:
        n = sp.prevprime(n)
```
