## Pandigital Primes

We shall say that an ð‘›-digit number is pandigital if it makes use of all the digits 1 to ð‘› exactly once. For example, 2143 is a 4-digit pandigital and is also prime.

What is the largest pandigital prime to exist?

### Methodology

Building a function to check whether a number is pandigital or not is rather simple. Since we're looking for a pandigital prime, we can start at 1 billion and work backwards. However, attempting this method will take too long. To narrow down our search we can rule out groups of numbers. Since the pandigital numbers all have the same digit sum (for a specific size), we can first look at the digit sums for numbers of size n:

1 : 1, 2 : 3, 3 : 6, 4 : 10, 5 : 15, 6 : 21, 7 : 28, 8 : 36, 9 : 45

We can see any pandigital numbers of size 2, 3, 5, 6, 8, and 9 aren't going to be prime because their digit sum is divisible by 3 (and thus also divisible by three). To narrow our search we can start with the seven digit numbers. If the loop is too long, we can then move down to the four digit numbers.  

```python
# use sympy to find primes faster
import sympy as sp
```

```python
# function to check if pandigital
def pandigital(n:int) -> bool:
    '''
    In put for the function will always be a prime number,
    thus this function only has to check whether the number is
    pandigital.
    '''
    # convert n to a string
    str_n = str(n)
    # first check if 0 is there
    if '0' in str_n:
        return False
    for i in range(1, len(str_n)+1):
        if str(i) not in str_n:
            return False
    return True
```

```python
# start from largest prime less than upper limit for pandigital numbers 
# 1000000000 
n = sp.prevprime(10000000)

# loop to iterate through primes (going backwards)
while True:
    if pandigital(n):
        print(f'Largest n-digit pandigital prime: {n}')
        break
    else:
        n = sp.prevprime(n)
```
