## Power Digit Sum

$2^{15} = 32768$ and the sum of its digits is $3 + 2 + 7 + 6 + 8 = 26$. What is the sum of the digits of the number $2^1000$?

### Methodology 

Fortunately, python can store $2^{1000}$ in memory. Thus, once we compute the value we can convert the value into a string, convert the string into a list of digits (int) and then compute the sum. 

```python
# calculate 2^1000
num = 2**1000
```

```python
# convert num to string, then to a list
num = str(num)

digits = [int(digit) for digit in num]

# print sum of digits
print(f'Sum of digits: {sum(digits)}')
```
