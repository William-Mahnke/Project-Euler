## Smallest Multiple 

2520 is the smallest number that can be divided by each of the numbers from 1 to 10 without any remainder.

What is the smallest positive number that is evenly divisible by all of the numbers from 1 to 20?

### Methodology

Code isn't needed to solve this problem. We can look at the prime factorization of 1-20 and see what factors are necessary for the product. Trivially the product will include all of the prime numbers up to 20 (2, 3, 5, 7, 11, 13, 17, 19). To make sure the product is divisible by the composite numbers we have to check the largest exponent each prime number will need to satisfy. For everything except 2 and 3 the exponent will just be 1:

5 is enough for 10, 15, and 20. 7 is enough for 14. 11, 13, 17, and 19 aren't factors for any other number. 

For 2 the exponent will be 4 since we need 16 as a factor. for 3 the exponent will be 2 since we need 9 and 18 as factors. Thus the prime factorization of the product is: $2^4 \times 3^2 \times 5 \times 7 \times 11 \times 13 \times 17 \times 19 = 232792560$. Therefore, 232792560 is our answer.
