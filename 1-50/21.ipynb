## Amicable Numbers

Let d(n) be defined as the sum of the proper divisors of n (numbers less than n which divide evenly into n). If d(a) = b and d(b) = a where $a \neq b$, then a and b are an amicable pair and each of a and b are called amicable numbers. 

For example, the proper divisors of 220 are 1, 2, 4, 5, 10, 11, 20, 22, 44, 55, and 110; therefore d(220) = 284. The proper divisors of 284 are 1, 2, 4, 71, and 142; so d(284) = 220. 

Evaluate the sum of all amicable numbers under 10000.

### Methodology

Creating a function to sum the proper divisors is a good first step. After creating the function, we can iterate through the numbers up to 10000 to check for these pairs. To avoid adding duplicates, creating a set rather than a list ensures those duplicates don't factor into the sum later.

```python
# function for getting proper divisors of a number 
def divisors(n: int) -> set:
    '''
    Returns the sum of proper divisors of n
    '''
    # special case (which shouldn't matter anyway)
    if n == 1:
        return 0
    # create set divs for divisors
    # include 1 so we don't accidentally add n to set
    divs = {1}
    i = 2

    # loop through numbers up to sqrt(n), add divisors to set
    while i * i <= n:
        if n % i == 0: # divisor, add to divs
            other = n / i # add other factor in pair
            divs.add(i)
            divs.add(other)
        i += 1

    # return the set
    return sum(divs)
```

```python
# list for amicable numbers, dictionary for number-sum pairs
amicables = set()

# loop through numbers
for i in range(2, 10000):
    # get sum of divisors for i
    val = divisors(i)
    if (divisors(val) == i) and (i != val): # amicable pair found
        amicables.add(i)
        amicables.add(val)
```

```python
# show sum of amicables
print(f'Sum of amicable numbers below 10000: {int(sum(amicables))}')
```
