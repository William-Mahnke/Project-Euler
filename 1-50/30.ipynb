## Digit Fifth Powers

Surprisingly there are only three numbers that can be written as the sum of fourth powers of their digits:

$1634 = 1^4 + 6^4 + 3^4 + 4^4$

$8208 = 8^4 + 2^4 + 0^4 + 8^4$

$9474 = 9^4 + 4^4 + 7^4 + 4^4$

As $1^4 = 1$ isn't a sum, it's not included.

The sum of these numbers is 1634 + 8208 + 9474 = 19316.

Find the sum of all the numbers that can be written as the sum of fifth powers of their digits.

### Methodology

First we can create a list mapping digits to their fifth power to save an exponent evaluation for each digit. Using this, we can iterate through numbers (starting after 1) and add any numbers to the total if they meet the criterion. As for the stopping point of this function, we can set an artificial upper bound and increase it if we get the wrong answer (the sum doesn't include all of the numbers yet).

```python
# list of fifth powers
# fifth[i] = i ** 5
fifth = {str(i) : i ** 5 for i in range(10)}
```

```python
# start variable for total and start loop
total = 0
for n in range(2, 1000000):
    if sum([fifth[d] for d in str(n)]) == n:
        print(n) # just to see the numbers (sanity check)
        total += n

# print total
print(f'Total: {total}')
```
