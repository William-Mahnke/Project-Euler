## Multiples of 3 or 5

If we list all the natural numbers below 10 which are multiples of 3 or 5, we get 3, 5, 6, and 9. The sum of these multiples is 23. Find the sum of all the multiples of 3 or 5 below 1000. 

### Methodology

Using modulus' of numbers, we can loop through all numbers below a thousand and add those multiples to a total. Another option is creating a set of the multiples for three and five and summing those.

#### First Method: Looping and Adding Multiples

```python
# iterate through numbers and only those which are multiples of three or five (or both)
total = 0

for i in range(1000):
    if (i % 3 == 0) or (i % 5 == 0):
        total += i

print(f"Sum of multiples of three and five below 1000: {total}")
```
