## Non-Abudant Sums

A perfect number is a number for which the sum of its proper divisors is exactly equal to the number. For example, the sum of the proper divisors of 28 would be $1 + 2 + 4 + 7 + 14 = 28$.

A number n is called deficient if the sum of its proper divisors is less than ð‘› and it is called abundant if this sum exceeds n.

As 12 is the smallest abundant number, $1 + 2 + 3 + 4 + 6 = 16$, the smallest number that can be written as the sum of two abundant numbers is 24. By mathematical analysis, it can be shown that all integers greater than 28123 can be written as the sum of two abundant numbers. However, this upper limit cannot be reduced any further by analysis even though it is known that the greatest number that cannot be expressed as the sum of two abundant numbers is less than this limit.

Find the sum of all the positive integers which cannot be written as the sum of two abundant numbers.

### Methodology



```python
# sqrt for iterating up to sqrt of n for factors
from math import sqrt, floor
```

```python
# function to determine if a number is abundant or not
def abundant(n: int) -> int:
    '''
    Finds divisors of a number and then checks if the number is abundant
    '''
    # iterate up to sqrt of n to find factor pairs
    factors = {1}
    for i in range(2, floor(sqrt(n))+1):
        if n % i == 0: # factor of n
            factors.add(i)
            factors.add(n/i)
    # sum set after iteration
    return sum(factors) > n
```

```python
# create list for abundant numbers
abundant_nums = []
# generate list of abundant numbers less than 28124
for i in range(1, 21824):
    if abundant(i):
        abundant_nums.append(i)
```

```python
# generate set of numbers which are the sum of two
# abundant numbers 
total = 0
for i in range(1, 28124):
    for num in abundant_nums:
        summable = False
        if (i - num) in abundant_nums:
            summable = True
            break
    if not summable:
        total += i
# print total 
print(f"Sum of numbers which aren't the sum of two abundant numbers: {total}")
```

```python
# sum pairs of abundant numbers and add to a set
summables = {2 * abundant_nums[-1]}
for i in range(0, len(abundant_nums)-1):
    for j in range(i , len(abundant_nums)):
        summables.add(abundant_nums[i] + abundant_nums[j])
```

```python
# use difference of sets to get set of numbers which aren't the
# sum of two abundant numbers 
non_summables = set(range(1, 28124)) - summables
```

```python
# sum elements of non_summables
print(f"Sum of numbers not the sum of two abundant numbers: {sum(non_summables)}")
```
