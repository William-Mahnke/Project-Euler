## Stealthy Numbers

A positive integer N is stealthy, if there exist positive integers a, b, c, d such that $ab = cd = N$ and $a + b = c + d + 1$. For example, $36 = 4 \times 9 = 6 \times 6$ is stealthy. 

You are also given that there are 2851 stealthy numbers not exceeding $10^6$. 

How many stealthy numbers are there that don't exceed $10^14$. 

```python
from math import sqrt
import numpy as np
```

```python
# function to check if a number is stealthy
def stealthy(n: int) -> bool:
    # first have to find all factor pairs of n
    factors = [(1, n)]
    i = 2
    while i < sqrt(n)+1:
        if n % i == 0:
            factors.append((i, n/i))
        i += 1
    # create set of sum of factors
    sums = np.array([sum(pair) for pair in factors])
    # create matrix of differences of the sums
    matrix = sums - sums.reshape(-1,1)
    return 1 in matrix
```

```python
stealthy(36)
```
