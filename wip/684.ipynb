## Inverse Digit Sum

Define $s(n)$ to be the smallest number that has a digit sum of n. For example s(10) = 19. Let $S(k) = \sum_{n=1}^k s(n)$. You are given $S(20) = 1074$. 

Further let $f_i$ be the Fibonacci sequence defined by $f_0 = 0, f_1 = 1$ and $f_i = f_{i-2} + f_{i-1}$ for all $i \geq 2$. 

Find $\sum_{i=2}^{90}S(f_i)$. Give your answer modulo 1000000007. 

```python
# create fibonacci sequence 
first = 0
second = 1
idx = 2
fib = []

while idx <= 90:
    val = first + second
    fib.append(val)
    first = second
    second = val
    idx += 1
```

```python
# show fibonacci sequence
print(fib)
```

```python
# create function to find smallest number with digit sum n
def smallest(digit_sum:int) -> int:
    '''
    The smallest number created for a given digit sum is
    a number with as many trailing nines after digit_sum mod 9
    (i.e. keep adding 9s at the end til you're left with a number between 1-8,
    then add that digit to the front of the number)
    '''
    nines = digit_sum // 9
    lead = digit_sum % 9
    num_str = str(lead) + nines * str('9')
    return int(num_str) 
```

```python
# function for calculating S values
def S(n: int) -> int:
    '''
    Takes a positive integer and returns sum{i = 1: n} s(n) 
    where s(n) is the smallest integer with a digit sum of n
    '''
    # first check is n is multiple of 9
    if n % 9 == 0:
        # define cap level
        cap = n // 9
        # can sum through the levels, not extra numbers to add
        part_one = sum([45 * (10 ** i) for i in range(cap)])
        part_two = sum([int('9' * i) * 9 for i in range(1, cap)])
        return part_one + part_two
    else:
        # first recursively call function to calculate sum up to nearest 
        # multiple of 9 less than n
        nearest_multiple = (n // 9) * 9
        part_one = S(nearest_multiple)
        # then calculate sum of remaining numbers using smallest function
        remaining_nums = list(range(nearest_multiple + 1, n + 1))
        part_two = sum(smallest(num) for num in remaining_nums)
        return part_one + part_two
```

```python
# calculate S(*) for fibonacci numbers
total = sum(S(num) for num in fib[:23])
print("Answer: {}".format(total % 1000000007))
```
