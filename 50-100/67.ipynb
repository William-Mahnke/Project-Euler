## Maximum Pathway Sum II

By starting at the top of the triangle below and moving to adjacent numbers on the row below, the maximum total from top to bottom is 23.

<center>
    <strong> 3 </strong> 
    <br>
    <strong> 7 </strong> 4
    <br> 
    2 <strong> 4 </strong> 6 
    <br>
    8 5 <strong> 9 </strong> 3
</center>

That is, 3 + 7 + 4 + 9 = 23.

Find the maximum total from top to bottom in `0067_triangle.txt`, a 15K text file containing a triangle with one-hundred rows.

### Method

From problem 18, we figured out an efficient algorithm for solving this kind of problem is summing the possible paths in the last two rows and only saving the larger of the two. Thus we can apply this process to this pyramid, with only slightly more work to convert the text file into a list of lists we can use to find the solution.

```python
# function to add possible paths for last two rows of triangle
def calculate_paths(higher: list, lower: list) -> list:
    '''
    Calculates the sum of the possible paths between two lists
    list1:list - the shorter of the input lists, aka the row above the 
    other row in the pyramid
    list2:list - the longer of the input lists (by one element),
    aka the row lower in the pyramid
    '''
    output_list = []
    for i in range(0, len(higher)):
        output_list.append(max(higher[i] + lower[i], higher[i] + lower[i+1]))
    return output_list
```

```python
# create empty list for pyramid 
pyramid = []
# open and read triangle file
with open("/Users/williammahnke/Desktop/Personal Coding/PyTorch Blitz/Project-Euler/files/0067_triangle.txt") as file:
    for line in file:
        lst = line[:-1].split(' ')
        level = [int(num) for num in lst]
        pyramid.append(level)
```

```python
# start path_sums with the sums of the last two layers
path_sums = calculate_paths(pyramid[98], pyramid[99])
# loop backwards through pyramid tiers and calculate path sums
for i in range(97, -1, -1):
    temp = calculate_paths(pyramid[i], path_sums)
    path_sums = temp
print(f'Maximum Path Sum: {path_sums[0]}')
```
