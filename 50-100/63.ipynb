## Powerful Digit Counts

The 5-digit number, $16807 = 7^5$, is also a fifth power. Similarly, the 9-digit number, $134217728 = 8^9$, is a ninth power.

How many n-digit positive integers exist which are also an nth power?

### Methodology 

This problem can be done manually. For each value of n, the only bases which will satisfy the condition are less than 10. Then, for each n, $10^n$ will have n+1 digits and every subsequent base used will result in a number with more than n digits. Thus for each n we only have to consider 1-9. Additionally, due to the nature of exponents, for each exponent n, the smallest base which satisfies the condition will be the lower bound for the smallest base for n+1. Therefore we can keep iterating through values of n until this lower bound becomes 10 (as in $9^n$ no longer satisfies the condition). While performing these operations we can simply just count every time the condition is true. 

```python
# initalize some variables 
count = 0
exp = 1
start = 1
# check for condition until starting base is 10
while start < 10:
    bases = list(range(start, 10))
    for b in bases:
        if len(str(b ** exp)) < exp:
            start += 1
        if len(str(b ** exp)) == exp:
            count += 1
    exp += 1
# print result
print(f'Number of times condition is met: {count}')
```
