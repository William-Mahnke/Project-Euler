## Hybrid Integers (WIP)

An integer of the form $p^qq^p$ with prime numbers $p /neq q$ is called a *hybrid-integer*. For example, $800 = 2^55^2$ is a hybrid-integer.

We define $C(n)$ to be the number of hybrid-integers less than or equal to n. You're given $C(800) = 2$ and $C(800) = 10790$.

Find $C(800800^{800800})$.

```python
# import log from math package
from math import log10
import sympy as sp
```

```python
# save 800800log(800800) as a variable so program calculates it once
# limit = 800800 * log10(800800)
limit = 800*log10(800)

# start with 2 for p
# iterate through higher and higher primes for q until limit is reached
p = 2
q = 3
hybrids = set()
iter = 0

while True:
    # first check if hybrid number would be less than limit
    if q * log10(p) + p * log10(q) > limit:
        # if we pass the limit, iterate p to next prime and reset q
        if iter == 0: # p is so large, any prime for q won't work, so we're done
            break
        else:
            p = sp.nextprime(p) # p goes to next prime up
            q = sp.nextprime(p) # to avoid repeats, start q at prime right after p

    # if limit isn't reached yet, calculate hybrid integers
    hybrids.add((p ** q) * (q ** p))

    # increment q to next prime, p stays the same until q has been exhausted
    q = sp.nextprime(q)
```
