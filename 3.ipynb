## Largest Prime Factor

The prime factors of 13195 are 5, 7, 13, and 29. What is the largest prime factor of 600851475143?

### Methodology

To find factors of the number, we only have to go up to the largest integer smaller than the square root of the number? Then keeping track of the prime factors of the number after dividing, we can iterate through all of the prime numbers and then extract the largest factor. 

```python
# libraries
from math import sqrt
```

```python
# create set for primes and start with 3 
primes = [2]
value = 3

num = 600851475143

while value < sqrt(num):
    isPrime = True
    # first check if value is prime
    for k in primes:
        if value % k == 0: # if value is divisible by a prime, then it's not prime itself
            isPrime = False
            value += 2 # can add two to value to skip the even numbers
    if isPrime: # checking if value is a factor of num
        primes.append(value)
        if num % value == 0: 
            print(value)
            num //= value
    value += 2
```
